```python
def printArray(L, start, end):
    for i in range(start, end + 1):
        print(L[i], end=' ')
    print()

def partitionQuickSort(L, start, end):
    pivot_value = L[end]
    left = start
    right = end - 1
    
    print(f"\n=== Partition 시작 (pivot: {pivot_value}, start: {start}, end: {end}) ===")
    print("초기 리스트 상태:", end=' ')
    printArray(L, start, end)
    
    while left <= right:
        # 왼쪽 포인터가 피벗보다 큰 값을 찾을 때까지 이동
        while left <= right and L[left] < pivot_value:
            left += 1
        
        # 오른쪽 포인터가 피벗보다 작은 값을 찾을 때까지 이동
        while left <= right and L[right] >= pivot_value:
            right -= 1
        
        if left < right:
            # 교환 전 상태 출력
            print(f"\n[교환 전] left: {left}, right: {right}")
            print("교환 전 리스트 상태:", end=' ')
            printArray(L, start, end)
            
            # 교환 진행
            L[left], L[right] = L[right], L[left]
            
            # 교환 후 상태 출력
            print(f"[교환 후] left: {left}, right: {right}")
            print("교환 후 리스트 상태:", end=' ')
            printArray(L, start, end)
    
    # 피벗 위치와 left 위치 교환
    L[left], L[end] = L[end], L[left]
    print(f"\n[피벗 교환] 최종 pivot 위치: {left}")
    print("피벗 교환 후 리스트 상태:", end=' ')
    printArray(L, start, end)
    print("=== Partition 종료 ===\n")
    
    return left

def quick_sort(L, start, end):
    if start < end:
        pivot = partitionQuickSort(L, start, end)
        quick_sort(L, start, pivot - 1)
        quick_sort(L, pivot + 1, end)

# 테스트 리스트
L = [90, 85, 25, 37, 61, 54, 11, 42]

print("퀵 정렬 실행 전 리스트 상태:", end=' ')
printArray(L, 0, len(L) - 1)

quick_sort(L, 0, len(L) - 1)

print("퀵 정렬 결과 리스트 상태:", end=' ')
printArray(L, 0, len(L) - 1)
```
